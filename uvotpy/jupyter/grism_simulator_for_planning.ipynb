```python
# this notebook shows how I use my Swift UVOT grism planning tool
from astropy.time import Time
from astropy import coordinates as coord
from uvotpy import grism_simulator_for_planning as sim

t1 = Time('2025-07-10')
pos = coord.get_icrs_coordinates('WD1057+719')

# the filter selection parameter is called 'wheelpos'
# values are :
# wheelpos = 160  for the UV clocked grism
# wheelpos = 200  for the UV nominal grism
# wheelpos = 955  for the visual grism clocked position
# wheelpos = 1000 for the visual grism nominal position

# to run the planning tool: 

S = sim.SimGrism(pos,datetime = t1, blim=16., chatter=1 ,offset=[0,0.], wheelpos=1000, roll=246.)

# blim is the limiting V magnitude for the stars considered in the plot. 
# offset [0.,0.] (with dots) is the center of the detector. 
# offsets are measured in arcminutes. 
# A useful offset for clocked uv observations is [0.,5.6] which places the spectrum 
# in the top middle of the detector and gets split off of the second order 
# so less overlap of first and second order
```
